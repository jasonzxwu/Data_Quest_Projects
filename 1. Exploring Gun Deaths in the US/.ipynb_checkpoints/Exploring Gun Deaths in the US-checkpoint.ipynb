# Exploring Gun Deaths in the US
The data set is obtained from fivethirtyeight. See [here](https://github.com/fivethirtyeight/guns-data) <br>

**Data Description**<br>

- year -- the year in which the fatality occurred.
- month -- the month in which the fatality occurred.
- intent -- the intent of the perpetrator of the crime. This can be Suicide, Accidental, NA, Homicide, or Undetermined.
- police -- whether a police officer was involved with the shooting. Either 0 (false) or 1 (true).
- sex -- the gender of the victim. Either M or F.
- age -- the age of the victim.
- race -- the race of the victim. Either Asian/Pacific Islander, Native American/Native Alaskan, Black, Hispanic, or White.
- hispanic -- a code indicating the Hispanic origin of the victim.
- place -- where the shooting occurred. Has several categories, which you're encouraged to explore on your own.
- education -- educational status of the victim. Can be one of the following:

    - 1 -- Less than High School
    - 2 -- Graduated from High School or equivalent
    - 3 -- Some College
    - 4 -- At least graduated from College
    - 5 -- Not available



```python
import csv
import datetime
import pprint
```

```python
data = list(csv.reader(open('guns.csv')))
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint (data[:5])
```

```python
headers = data[0]
data=data[1:]
```

```python
pp.pprint(data[:5])
```

```python
#use list comprehension to extract year column from data
years = [row[1] for row in data]
```

```python
#create empty dictionary 
year_counts={}

#extract year counts

for year in years:
    if year in year_counts:
        year_counts[year] +=1
    else:
        year_counts[year] =1 
```

```python
year_counts
```

```python
#use list comprehension to extract dates
dates = [datetime.datetime(year=int(row[1]), month=int(row[2]), day=1) for row in data]
```

```python
dates[:5]
```

```python
date_counts={}

#extract day counts

for date in dates:
    if date in date_counts:
        date_counts[date] +=1
    else:
        date_counts[date] =1 
```

```python
date_counts
```

We can filter our results, and restrict them to the `Homicide` intent. This will tell us what the gun-related murder rate per `100000` people in each racial category is. In order to do this, we'll need to redo our work in generating `race_counts`, but only count rows where the `intent` was `Homicide`.

We can do this by first extracting the `intent` column, then using the *enumerate()* function to loop through each index and value in the race column. If the value in the same position in `intents` is `omicide`, we'll count the value in the race column.

Finally, we'll use the `mapping` dictionary to convert from raw counts to rates.

```python
headers.index('intent')
```

```python
intents = [row[3] for row in data]
```

```python
races = [row[7] for row in data]
```

```python
homicide_race_counts={}

for i, v in enumerate(races):
    if intents[i] =='Homicide':
        if v not in homicide_race_counts:
            homicide_race_counts[v]=1
        else:
            homicide_race_counts[v] +=1
            
```

```python
homicide_race_counts
```

```python
for race in homicide_race_counts:
    homicide_race_counts[race] = (homicide_race_counts[race]/ mapping[race])*(10**5)
```

```python
homicide_race_counts
```

## Explore the link between month and homicide rate

```python
print (headers)
```

```python
months = [int(row[2]) for row in data]
```

```python
months
```

```python
homicide_month_counts={}
for i,m in enumerate(months):
    if intents[i]=='Homicide':
        if m not in homicide_month_counts:
            homicide_month_counts[m] =1
        else:
            homicide_month_counts[m] +=1
```

```python
homicide_month_counts
```

```python
homicide_month_df = pd.DataFrame.from_dict(homicide_month_counts,orient='index')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
homicide_month_df.reset_index(inplace=True)
```

```python
homicide_month_df.columns=['month','homicide_counts_by_month']
```

```python
homicide_month_df
```

```python
homicide_month_df.plot(x='month',y='homicide_counts_by_month',kind='scatter',
                      title='Monthly number of homicides from 2012 to 2014',
                  figsize=(10,6))
```
